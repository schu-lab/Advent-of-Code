```python
def visualize(array):
    color_mapping = {
        '*': '\033[31m',  # Red
        '#': '\033[32m',  # Green
        '+': '\033[33m',  # Yellow
        '$': '\033[34m',  # Blue
    }
    reset_color = '\033[0m' # Reset color to default

    for row in array:
        for item in row:
            if str(item) in color_mapping:
                color_code = color_mapping.get(str(item))
                print(f"{color_code}{item}{reset_color}", end = ' ')
            else:
                print(f"{item}{reset_color}", end = ' ')
        print()

schematic = parse_schematic(data)
visualize(schematic)
```
