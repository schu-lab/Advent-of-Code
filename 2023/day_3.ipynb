```python
def parse_schematic(data):
    schematic = []
    for line in data:
        columns = []
        row = line.strip()
        for character in row:
            columns.append(character)
        schematic.append(columns)
    return schematic

# # Check Symbols:
# def check_symbol(schematic, check_row, check_column):
#     symbols = ['*', '#', '+', '$', '/']
#     check_value = schematic[check_row][check_column]
#     if check_value in symbols:
#         print("Symbols:  " + check_value + "  ...found in...  [Row: " + str(check_row)
#               + "] [Column: " + str(check_column) + "]" )
#         status_symbol = True
#     else:
#         status_symbol = False
#     return status_symbol

# # Check Numbers:
# def check_number(schematic, check_row, check_column):
#     check_value = schematic[check_row][check_column]
#     if check_value.isdigit():
#         print("Number:  " + check_value + "  ...found in...  [Row: " + str(check_row)
#               + "] [Column: " + str(check_column) + "]" )
#         status_number = True
#     else:
#         status_number = False
#     return status_number

# Find Symbols
def parse_input(new_schematic):
    symbols = {}
    numbers = []
    for row, row_value in enumerate(data):
        number = ''
        for column, column_value in enumerate(row_value):
            # Find and store location of symbols (Symbol, Row, Column):
            if not column_value.isdigit() and column_value != '.' and column_value != '\n':
                symbols[row, column] = column_value
                # print(symbols)

            # Find and combine into number
            if column_value.isdigit():
                if not number:
                    column_start = column_value
                number += column_value
                # print(number)
            # Store location of numbers (Number, Row, Column(Start/End)
            elif number:
                column_start = column - len(number)
                column_end = column - 1
                numbers.append((int(number), row, column_start, column_end))
                # print(numbers)
                number = ''
    return symbols, numbers

def part_numbers(symbols, numbers):
    parts_observed = set()
    part_numbers = []
    # print(symbols)
    for row_symbol, column_symbol in symbols:
        # print("[Row: " + str(row_symbol) + "] [Column: " + str(column_symbol) + "]")
        for i, (number, row_num, column_start_num, column_end_num) in enumerate(numbers):
            # print(number, row_num, column_start_num, column_end_num)
            symbol_row_lookup = ((row_num - 1) <= row_symbol <= (row_num + 1))
            symbol_column_lookup = ((column_start_num -1) <= column_symbol <= (column_end_num + 1))
            if symbol_row_lookup and symbol_column_lookup:
                # print("Symbol Row: " + str(row_symbol) + " + Symbol Column: " + str(column_symbol))
                parts_observed.add(i)
                # print("i: " + str(i))
                # print(parts_observed)
                part_numbers.append(number)
                # print(part_numbers)
    return part_numbers

schematic = parse_schematic(data)
symbols, numbers = parse_input(schematic)
part_number_list = part_numbers(symbols, numbers)

# for i in range(len(symbols)):
#     print("# of Symbols: " + str(i))
# for i in range(len(numbers)):
#     print("# of Part Numbers: " + str(i))

total_sum = sum(part_number_list)
print("Sum of all part numbers in engine schematic: ", total_sum)
```

```python
def visualize(array):

    gear_color = '\033[31m'
    symbol_color = '\033[32m'
    reset_color = '\033[0m'

    for row in array:
        for item in row:
            if item == '*':
                print(f"{gear_color}{item}{reset_color}", end = '')
            elif not item.isdigit() and item != '.':
                print(f"{symbol_color}{item}{reset_color}", end = '')
            else:
                print(f"{item}{reset_color}", end = '')
        print()

schematic = parse_schematic(data)
visualize(schematic)
```
