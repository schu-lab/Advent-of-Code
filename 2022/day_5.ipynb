```python
"""
Author: S. Chu
Date:

--- Day 5: Supply Stacks ---
The expedition can depart as soon as the final supplies have been unloaded from the ships. Supplies are stored in stacks of marked crates, but because the needed supplies are buried under many other crates, the crates need to be rearranged.

The ship has a giant cargo crane capable of moving crates between stacks. To ensure none of the crates get crushed or fall over, the crane operator will rearrange them in a series of carefully-planned steps. After the crates are rearranged, the desired crates will be at the top of each stack.

The Elves don't want to interrupt the crane operator during this delicate procedure, but they forgot to ask her which crate will end up where, and they want to be ready to unload them as soon as possible so they can embark.

They do, however, have a drawing of the starting stacks of crates and the rearrangement procedure (your puzzle input). For example:

    [D]
[N] [C]
[Z] [M] [P]
 1   2   3

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
In this example, there are three stacks of crates. Stack 1 contains two crates: crate Z is on the bottom, and crate N is on top. Stack 2 contains three crates; from bottom to top, they are crates M, C, and D. Finally, stack 3 contains a single crate, P.

Then, the rearrangement procedure is given. In each step of the procedure, a quantity of crates is moved from one stack to a different stack. In the first step of the above rearrangement procedure, one crate is moved from stack 2 to stack 1, resulting in this configuration:

[D]
[N] [C]
[Z] [M] [P]
 1   2   3
In the second step, three crates are moved from stack 1 to stack 3. Crates are moved one at a time, so the first crate to be moved (D) ends up below the second and third crates:

        [Z]
        [N]
    [C] [D]
    [M] [P]
 1   2   3
Then, both crates are moved from stack 2 to stack 1. Again, because crates are moved one at a time, crate C ends up below crate M:

        [Z]
        [N]
[M]     [D]
[C]     [P]
 1   2   3
Finally, one crate is moved from stack 1 to stack 2:

        [Z]
        [N]
        [D]
[C] [M] [P]
 1   2   3
The Elves just need to know which crate will end up on top of each stack; in this example, the top crates are C in stack 1, M in stack 2, and Z in stack 3, so you should combine these together and give the Elves the message CMZ.

After the rearrangement procedure completes, what crate ends up on top of each stack?
"""
```

```python
# Open file and check:
filename = 'day_5.txt'

with open(filename) as fh:
    data = fh.readlines()
```

```python
# Function to determine which crate is on top

# Setup Stack in crate_data:
def stack_setup(data):
    i = 0
    while "move" not in data[i]:
        i += 1

    # Split Crate Data and Move Data:
    crate_data = data[:i]
    move_data = data[i:]
    # print(crate_data)
    # print(move_data)

    # # Display Crates:
    # for i in crate_data:
    #     print("Display Crate Data: " + str(i))

    # Setup number of stacks
    num_stacks = len(crate_data[-2].strip().split())

    stacks = []
    for i in range(num_stacks):
        stacks.append(list())
    # print(stacks)

    # Parse stacks
    for i in range(len(crate_data)-3, -1, -1):
        for j in range(num_stacks):
            ind = j * 4 + 1
            if ind < len(crate_data[i]):
                # print("Crate Data Row (i): " + str(i) + "; Crate Data Column (Ind): " + str(ind))
                # print("Selected Crate Data Item: " + crate_data[i][ind])
                if crate_data[i][ind] != " ":
                    stacks[j].append(crate_data[i][ind])
                    # print("Stack " + str(j + 1) + "-" + str(i + 1) + " :" + str(stacks[j]))

        # Display Stacks:
        # print("Display Stacks: " + str(stacks))
    return stacks

def move_setup(data):
    i = 0
    while "move" not in data[i]:
        i += 1

    # Split Crate Data and Move Data:
    crate_data = data[:i]
    move_data = data[i:]
    # print(crate_data)
    # print(move_data)

    # # Display Moves:
    # for i in move_data:
        # print("Display Move Data: " + str(i))

    # Parse Moves:
    move_command = []
    for item in move_data:
        # print("Command: " + str(item))
        junk, rem = item.strip().split("move ")
        # print("Split: " + rem)
        count, rem = rem.strip().split("from ")
        # print("Number of Crates: " + count + "; Move Stack: " + rem)
        stack_a, stack_b = rem.strip().split("to ")
        # print("Number of Crates: " + count + "; From: " + stack_a + "; To: " + stack_b)
        move_command.append([int(count), int(stack_a)-1, int(stack_b)-1])
        # print(move_command)       # Stack A and B are stored in array (N-1)
    return move_command

def current_move_commands(input):
    for row in input:
        print("Move Command - Number of Crates: " + str(row[0]) +
              "; From Stack #: " + str(row[1]+ 1) +
              "; To Stack #: " + str(row[2] + 1))

def current_stack_status(input):
    # Display Current:
    for i in range(len(input)):
        print("Stack #" + str(i + 1) + ": " + str(input[i]))

def move_crates(current_stacks, stack_a, stack_b):

    current_stacks[stack_b].append(current_stacks[stack_a].pop())

    # # Display Move Command:
    # print(current_stacks)
    # print("\nMove Command - From Stack: " + str(stack_a + 1)
    #       + "; To Stack: " + str(stack_b + 1))

    # # Display Current Stacks:
    # print("\nMove Status:")
    # current_stack_status(stacks)

def move_stacks(move_cmd, current_stacks):

    count, stack_a, stack_b = move_cmd
    for n in range(count):
        move_crates(current_stacks, stack_a, stack_b)
    # current_stack_status(stacks)

    # # Display Stacks/Move Command:
    # # print(current_stacks)
    # # print(move_cmd)
    # print("Number of Crates: " + str(count)
    #       + "; From Stack: " + str(stack_a + 1)
    #       + "; To Stack: " + str(stack_b + 1))

def top_crate_on_stacks(stacks):
    print("\nCurrent Status - Top Crate on Stacks:")
    print(*[stack[-1] for stack in stacks], sep="")
```

```python
# Sequence for Crate Mover 9000:
# Move Crates one at a time

stacks = stack_setup(data)
moves = move_setup(data)
# current_move_commands(moves)
print("Original Stack Setup:")
current_stack_status(stacks)

for move in moves:
    move_stacks(move, stacks)

print("\nFinal Stack Setup:")
current_stack_status(stacks)
top_crate_on_stacks(stacks)
```
